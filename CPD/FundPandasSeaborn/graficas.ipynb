# Graficas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_csv('penguins.csv')
```

```python
sns.displot(archivo, x='flipper_length_mm')
plt.show()
```

```python
sns.displot(archivo, x='flipper_length_mm', hue='species')
plt.show()
```

```python
sns.displot(archivo, x='flipper_length_mm', hue='species', multiple='stack')
plt.show()
```

```python
sns.pairplot(archivo, hue='species')
plt.show()
```

```python
sns.boxplot(x='body_mass_g', y='sex', hue='species', data=archivo, palette=['m','g','b'])
plt.show()
```
