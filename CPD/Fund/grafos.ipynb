```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')

pos = {'A': (0, 0), 'B': (1, 0), 'C': (2, 3), 'D': (0, 5)}

G.add_edge('A', 'B')
G.add_edge('C', 'D')
G.add_edge('A', 'D')
G.add_edge('B', 'D')
```

```python
nx.draw_networkx(G, pos=pos)
ax = plt.gca()
ax.margins(0.20)
plt.axis('off')
plt.show()
```

```python
with open('from.txt', 'r') as file1, open('to.txt', 'r') as file2:
    aristas_from = [int(linea) for linea in file1]
    aristas_to = [int(linea) for linea in file2]
```

```python
G = nx.DiGraph()
G.add_nodes_from(range(1, 6))
G.add_edges_from(zip(aristas_from, aristas_to))
```

```python
nx.draw_networkx(G, pos=nx.spring_layout(G))
ax = plt.gca()
plt.axis('off')
plt.show()
```
