```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_csv('penguins.csv')
```

```python
sns.displot(archivo, x='flipper_length_mm')
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.displot(archivo, x='flipper_length_mm', hue='species')
plt.show()
```

```python
sns.displot(archivo, x='flipper_length_mm', hue='species', multiple='stack')
plt.show()
```

```python
sns.pairplot(archivo, hue='species')
plt.show()
```

```python
sns.boxplot(x='body_mass_g', y='sex', hue='species', data=archivo, palette=['m','g','b'])
plt.show()
```

```python
sns.scatterplot(data=archivo, x='body_mass_g', y='flipper_length_mm', hue='species', color='.15')
```

```python
sales_data = [23, 40, 14, 16, 50]
products = ['Hats', 'T-shirts', 'Pants', 'Jackets', 'Shoes']
explode = [0, 0, 0, 0, 0.1]
plt.pie(sales_data, labels=products, explode=explode, autopct='%1.2f%%')
plt.show()
```
