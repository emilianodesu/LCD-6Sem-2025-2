Manejo de valores nulos

```python
# Completa los valores faltantes de la columna bill_length_mm con la media de la columna
archivo['bill_length_mm'] = archivo['bill_length_mm'].fillna(archivo['bill_length_mm'].mean())
archivo
```

```python
# Completa los valores faltantes de la columna sex con la moda
archivo['sex'] = archivo['sex'].fillna(archivo['sex'].mode()[0])
archivo
```

```python
# Borrar registros
archivo = archivo.drop(2)
archivo = archivo.drop([0, 1, 3])
archivo
```

```python
# Eliminar filas con valores nulos en cualquier columna
archivo = archivo.dropna()
archivo
```

```python
# Eliminar filas con valores nulos solo en columnas específicas
archivo = archivo.dropna(subset=['sex', 'island'])
archivo
```

```python
# Crear columnas nuevas
archivo['body_mass_kg'] = archivo['body_mass_g'] / 1000
archivo
```

```python
archivo['gordito'] = np.where(archivo['body_mass_g'] > 4220, 1, 0)
archivo
```

Dummies

```python
dummies = pd.get_dummies(archivo['island'])
archivo = pd.concat([archivo.drop('island', axis=1), dummies], axis=1)
archivo
```
